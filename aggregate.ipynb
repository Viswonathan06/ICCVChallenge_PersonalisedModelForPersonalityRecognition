```python
comb_pred = pd.DataFrame()
# loop to combine all the files to later find mean. 
# Change loop variable 'files_combined' to whatever you want.
for file in files_combined:
    comb_pred = comb_pred.append(file)
    
```

```python
aggregate_pred = comb_pred.groupby(comb_pred.columns[0], as_index=False).mean()
aggregate_pred.head(20)
```
